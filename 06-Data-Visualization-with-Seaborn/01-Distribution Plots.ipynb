```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'],bins=40) #the blue line is called kde which you can set it false like kde=false 
#this is a distribution plot
```

```python
sns.jointplot(x='total_bill',y='tip', data=tips, kind='reg')
```

```python
sns.pairplot(data=tips, hue='sex', palette='coolwarm')
```

```python
sns.rugplot(data=tips['total_bill'])
```
