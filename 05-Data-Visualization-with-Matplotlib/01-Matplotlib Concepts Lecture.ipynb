```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL METHOD TO CREATE GRAPHS
plt.plot(x,y)
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Title")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Object-Oriented Method
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) #the list is [left,bottom,width,height]
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title("Larger Plot")

axes2.plot(y,x)
axes2.set_title("Smaller Plot")
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

#axes.plot(x,y)
#for curr_ax in axes: #the axes is the just the manual thing we created by fig.add_axes
#    curr_ax.plot(x,y)
#you can also iterate it meaning axes[0].plot(x,y)

axes[0].plot(x,y)
axes[0].set_title("First Plot")

axes[1].plot(y,x)
axes[1].set_title("Second Plot")
```

```python
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')
#Before you call this there needs to be label in what you plot
ax.legend(loc=0) #NOTE 0 means best location you can look up the predefined location other wise pass a tuple (x,y) for loc
```

# Plot Apperance 

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#FF8C00',linewidth=1,alpha=1,linestyle='--',marker='^',markersize='10',
        markerfacecolor='red',markeredgewidth='3',markeredgecolor='blue') #RGB HEX Code or pre-defined colour like blue,orange,green... also work, linewidth can also be lw= and alpha ranges from 0 to 1 and it is the transparcy linestyle can also be ls= and they are predefined
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',lw=2,ls='--')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
