# Style and Colour

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.set_style('ticks')
sns.countplot(data=tips, x='sex')
sns.despine()
```

```python
sns.set_context('poster')
sns.countplot(data=tips, x='sex')
```

```python
sns.set_context('notebook')
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='seismic')
```
