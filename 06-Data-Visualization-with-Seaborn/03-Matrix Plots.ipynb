# Matrix Plots

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
tc = tips.corr()
```

```python
sns.heatmap(data=tc, annot=True, cmap='coolwarm')
```

```python
fp = flights.pivot(index='month', columns='year', values='passengers')
```

```python
sns.heatmap(data=fp, cmap='magma', linecolor='white', linewidths=1)
```

```python
sns.clustermap(data=fp, standard_scale=1)
```
