```python
import numpy as np
```

```python
import seaborn as sns
```

```python
import pandas as pd
```

```python
%matplotlib inline
```

```python
df1 = pd.read_csv('df1', index_col=0)
df1.head()
```

```python
df2 = pd.read_csv('df2')
df2.head()
```

```python
df1['A'].hist(bins=30)
```

```python
df1['A'].plot(kind='hist', bins=30)
```

```python
df1['A'].plot.hist()
```

```python
df2.plot.area(alpha=0.4)
```

```python
df2.plot.bar() #could set stacked=True
```

```python
df1.plot.scatter(x='A', y='B', c='C', cmap='magma') #or u can do s=df1['C']*100
```

```python
df2.plot.box()
```
